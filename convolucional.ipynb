```python
from sklearn.datasets import load_sample_image
from google.colab.patches import cv2_imshow
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
from tensorflow import keras
```

```python
china = load_sample_image("china.jpg")
flower = load_sample_image("flower.jpg")
```

```python
cv2_imshow(china)
```

```python
cv2_imshow(flower)
```

```python
china_scale = china/255
flower_scale = flower/255
```

```python
images = np.array([flower_scale,china_scale])
batch_size, height, width, channels = images.shape
```

```python
filters = np.zeros((7,7,channels,2),dtype = np.float32)
filters[:,3,:,0] = 1 #Filtro de lineas verticales.
filters[3,:,:,1] = 1 #Filtro de lineas horizontales.
```

```python
outputs = tf.nn.conv2d(images, filters, strides=1, padding ='SAME')
outputs.shape
```

```python
plt.imshow(outputs[1,:,:,0], cmap='gray')
```
